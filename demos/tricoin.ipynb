```python
nsamples = 1000000  

# Function to recreate in our DSL
def test_cond():
    if coin(0.8) == 1:
        return 100
    else:
        return 0

# Pure sampling for 'if' function
prog = Dist(Sampler(lambda: test_cond()), nsamples)
print("Mean: ", prog.estimate(), "Variance: ", prog.variance().estimate())

# Using the 'If' construct.
cond = Dist(Sampler(lambda: coin(0.8)), nsamples)
if_expr = Exact(100)
else_expr = Exact(0)

prog2 = If(cond, if_expr, else_expr)
print("Mean: ", prog2.estimate(), "Variance: ", prog2.variance().estimate())

```
