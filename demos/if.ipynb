```python
nsamples = 1000000  

# Function to recreate in our DSL
def test_cond():
    if coin(0.8) == 1:
        return 100
    else:
        return 0

coin_callable = NamedCallable(lambda: coin(0.8), "coin")

# Pure sampling for 'if' function
callable = NamedCallable(lambda: test_cond(), "test_cond")
prog = Dist(Sampler(callable), nsamples)
print("Mean: ", prog.estimate(), "Variance: ", prog.variance().estimate())

# Using the 'If' construct.
cond = Dist(Sampler(coin_callable), nsamples)
if_expr = Exact(100)
else_expr = Exact(0)

prog2 = Profile(Add(If(cond, if_expr, else_expr), Sampler(coin_callable)))
print(pretty(prog2))

print("Mean: ", prog2.estimate(), "Variance: ", prog2.variance().estimate())
print("Summary:\n", prog2.summary())

```
